```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
import seaborn as sns; sns.set(color_codes=True)
```

```python
sim = 'simu_update_'
seq = 'seq_update_'
```

# Sequential update

```python
d_plot = []
for i in range(1,6):
    seq_file = seq+str(i)+'/score.pickle'
    f = open(seq_file, 'rb') 
    socre_list = pickle.load(f)[0:19000]
    d_plot.append(socre_list)
    
    
font = {'family' : 'normal', 'size' : 22}
fig=plt.figure(figsize=(18, 4), dpi= 80, facecolor='w', edgecolor='k')
plt.rc('font', **font)
plt.rc('axes', titlesize=18)
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
data = np.array(d_plot)
fig=plt.figure(figsize=(18, 4), dpi= 80, facecolor='w', edgecolor='k')
plt.subplot(1,2,1)
ax = sns.tsplot(data=data, ci = [68,95])
plt.xlabel('episodes',fontsize=19)
plt.ylabel('Adversary rewards',fontsize=19)
plt.subplot(1,2,2)
ax = sns.tsplot(data=-data, ci = [68,95])
plt.xlabel('episodes',fontsize=19)
plt.ylabel('Observer rewards',fontsize=19)
fig.savefig('seq.pdf', bbox_inches='tight')
plt.show()
```
